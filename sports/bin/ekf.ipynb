# Homography estimation for single pan-tilt camera video though extend Kalman filter
### Mapping the court template to video frame
By pre-process the input video frame, we first obtain the boundary points.
We will use one frame as an example in here.

```python
frame = cv2.imread("test0.png")
tpl = cv2.imread("tpl.png", 0)
tplPts = cv2.findNonZero(cv2.bitwise_not(tpl)).reshape(-1, 2)
H, W, _ = frame.shape
fig, (ax1, ax2) = plt.subplots(1,2)
ax1.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
ax2.imshow(tpl, cmap="gray")
```

```python
frameHSV = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV) # change color space to hsv for filtering color channel 
threshColor = cv2.inRange(frameHSV, np.array([0, 47, 151]), np.array([16, 255, 255]))
threshColor = cv2.morphologyEx(threshColor, cv2.MORPH_OPEN, np.ones((3, 3), np.uint8))
edges = cv2.Canny(threshColor, 200, 255, apertureSize=3) # calculate canny edge 
edges[565:650, 240:950] = 0
cv2.circle(edges, (1042, 620), 29, 0, -1)  # remove the icon on screen
edgesPts = cv2.findNonZero(edges).reshape(-1, 2) # collect all boundary points
fig, (ax1, ax2) = plt.subplots(1,2)
ax1.imshow(threshColor, cmap="gray")
ax2.scatter(edgesPts[:,0], edgesPts[:,1], s=5, c='red', marker='.')
ax2.invert_yaxis(); ax2.set_aspect('equal')
ax2.set_xlim([0,1280]); ax2.set_ylim([720, 0])
```

Notice that `edgesPts` is the boundary point sets of the given frame.
Obviously, we see quite a lot noises or outliers as well as the main court skeleton on `edgesPts`. 
Now we are trying to match `edgesPts` along with our court template `tpl`. 
We know both of them are point of sets, but we do not know their correspondent relationships.
Let's assume we have a fairly good initial position, then their relationships can be estimated with their nearset neighbours in Eudldian distance.

```python
courtPts = [[613, 161], [613, 770], [394, 460], [835, 460]]
framePts = [[619, 260], [626, 720], [58, 437], [1175, 424]]
M = cv2.getPerspectiveTransform(np.float32(courtPts), np.float32(framePts)) # initial guess
cRot = cv2.warpPerspective(tpl, M, (W, H))
cRotPts = cv2.findNonZero(cRot).reshape(-1,2)
fig, (ax1, ax2) = plt.subplots(1,2)
ax1.imshow(cRot, cmap="gray")
ax2.scatter(edgesPts[:,0], edgesPts[:,1], s=5, c='r', marker='.')
ax2.invert_yaxis(); ax2.set_aspect('equal')
ax2.set_xlim([0,1280]); ax2.set_ylim([720, 0])
```

```python
nbrs = NearestNeighbors(n_neighbors=1, radius=1.0).fit(edgesPts) # index all the boundary points 
distances, indices = nbrs.kneighbors(cRotPts) # match every court pts to the index
idx = distances < 7 # only select pts within 5 radius
matchedEdge = edgesPts[indices[idx]]
matchedCourt = cRotPts[idx.ravel()]
```

```python
fig, (ax1, ax2) = plt.subplots(1,2)
org = frame.copy()
org[..., 1] = cv2.bitwise_or(cRot, org[...,1])
ax1.imshow(cv2.cvtColor(org, cv2.COLOR_BGR2RGB))
#ax1.scatter(edgesPts[:,0], edgesPts[:,1], s=5, c='r', marker='.')
ax1.invert_yaxis(); ax1.set_aspect('equal')
ax1.set_xlim([0,1280]); ax1.set_ylim([720, 0])
ax2.scatter(matchedCourt[:,0], matchedCourt[:,1], s=5, c='r', marker='.')
ax2.invert_yaxis(); ax2.set_aspect('equal')
ax2.set_xlim([0,1280]); ax2.set_ylim([720, 0])
```

## Direct linear transformation (DLT)
From the initial guess, we see the matched points are fairly good, but not perfect. 
Based on the frame and template points sets and their reltionship, we could use DLT method to improve our estimation.
Formally, the points on court will be mapped on the frame based on perspective tranformation $\mathbf{M}$.
By represent the points in homogenous form, for each pairs of points, we could have,

$$
\begin{bmatrix} x'\\y'\\1 \end{bmatrix} = 
\begin{bmatrix} 
H_{11} & H_{12} & H_{13} \\
H_{21} & H_{22} & H_{23} \\
H_{31} & H_{32} & 1 
\end{bmatrix}
\begin{bmatrix} x\\y\\1 \end{bmatrix}
$$

which implies that, 
$$
x' = \frac{H_{11}x + H_{12}y + H_{13}}{H_{31}x + H_{32}y + 1} \\
y' = \frac{H_{21}x + H_{22}y + H_{23}}{H_{31}x + H_{32}y + 1}
$$

Rearrange to, 
$$
\begin{bmatrix} x'\\y' \end{bmatrix} =
\begin{bmatrix} 
x & y & 1 & 0 & 0 & 0 & -x x' && -y x' \\
0 & 0 & 0 & x & y & 1 & -x y' && -y y' 
\end{bmatrix}
\begin{bmatrix} 
H_{00} \\ H_{01} \\ H_{02} \\
H_{10} \\ H_{11} \\ H_{12} \\
H_{20} \\ H_{21} \\
\end{bmatrix} \\
\mathbf{p}' = \mathbf{A} \mathbf{h}
$$

The above equation have infinite solution due to only one point pair is considered.
In realitity, we collect lots of $(x,y) - (x',y')$ with substantial noise and outliers. 
Thus, without considering other knowledges, we could estimate the $\mathbf{H}$ through ordinary least squares method. 
$$
\hat{\mathbf{h}} = \mathbf{A}^{-1} \mathbf{p}'
$$

In opencv, `cv2.findHomography()` can easily do the above job without care too much about the linear equations.

```python
%%time
fig, axs = plt.subplots(5,5)
m = M
for ax in axs.ravel():
    cRot = cv2.warpPerspective(tpl, m, (W, H))
    cRotPts = cv2.findNonZero(cRot).reshape(-1,2)
    distances, indices = nbrs.kneighbors(cRotPts) 
    idx = distances < 10 
    matchedEdge = edgesPts[indices[idx]]
    matchedCourt = cRotPts[idx.ravel()]
    dh, ret = cv2.findHomography(matchedCourt, matchedEdge)
    m = np.dot(dh, m)
    img = frame.copy()
    img[..., 1] = cv2.bitwise_or(cRot, img[...,1])
    ax.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    ax.invert_yaxis(); ax.set_aspect('equal')
    ax.set_xlim([0,1280]); ax.set_ylim([720, 0])
    ax.set_title("matched # %s"%matchedCourt.shape[0])
    ax.axis('off')
```

Based on a reasonable initial $\mathbf{H}$, we obtain a good rectification result iteratively through DLT method.
The result is promising, but the converge speed is not good enough to deploy it on real-time application.
There are two factors greatly impact to the speed. One is the initial $\mathbf{H}$ selection, and the other is the DLT method itself. 
The first one can be addressed by considering the temporal realtionships in the video which we will covered it in later.
For the second one, we can first eliminate the number of points required in each DLT iteration. 
The dimensonal in boudary set for each frame are quite large. 
Hence bringing lots complexity in both nearest neighbours and matrix inversion in DLT.
To put it simply we could just sample partital points in the template set to reduce the time.


```python
%%time
fig, axs = plt.subplots(5,5)
m = M
for ax in axs.ravel():
    cRot = cv2.warpPerspective(tpl, m, (W, H))
    cRotPts = cv2.findNonZero(cRot).reshape(-1,2)
    rndIdx = np.random.choice(cRotPts.shape[0], 400)
    distances, indices = nbrs.kneighbors(cRotPts[rndIdx]) 
    idx = distances < 10 
    matchedEdge = edgesPts[indices[idx]]
    matchedCourt = cRotPts[rndIdx][idx.ravel()]
    dh, ret = cv2.findHomography(matchedCourt, matchedEdge)
    m = np.dot(dh, m)
    img = frame.copy()
    img[..., 1] = cv2.bitwise_or(cRot, img[...,1])
    ax.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    ax.invert_yaxis(); ax.set_aspect('equal')
    ax.set_xlim([0,1280]); ax.set_ylim([720, 0])
    ax.set_title("matched # %s"%matchedCourt.shape[0])
    ax.axis('off')
```

## Temporal Video Model 
Besides the single frame optimization, the temporal relationship in consective frames are also important to improve the homography estimation. 
Based on our discussion, the performance are highly leveraged by the initial homography guess. A poor selection will degrade the performance heavily and may even induce the wrong anwser. 
In order to select the best initial solution, we could leverage the additional information provided from the application. 
Specifically, for a single pat-tilt camera video, we know as long as the video are captured consectively, the frame will have temporal relationship. 
Furthermore, because of the characteristic of the pat-tilt-zoom camera, there essentially only have three degree of freedoms driven by pan, tilt and zoom.
The camera motion state can not be directly obtained from the video frame, all we know is their temporal relationship due to the physical restriction. 
We would like to build a hidden layer model to inference them, and then possibly lerverage this knowledge to improve our homogrphy. 

We define $dx$, $dy$ and $f$ as the pan, tilt and zoom to charactize the camera state $\mathbf{x}$ respectively. 
$$
\mathbf{x}_k = \begin{bmatrix} dx\\dy\\f \end{bmatrix}
$$
It is relatively easy and intuitive to model the state $\mathbf{x}$ as a constant velocity model due to the operation behind of the camera. 
But what is the relationship between $\mathbf{x}$ and the video frame?
We do know once the $\mathbf{x}$ change, the camera move it focus accrodingly, hence captured different picture we observed on a frame.

Let's review the camera model and perspective transformation first. 
And see if there are any model can put them together. 

The pin-hole camera model for the perspective pan-tilt-zoom camera consists of a mapping from 3D projective space to 2D projective space.
The perspective transformation matrix $\mathbf{P}$ can be decomposed as
$$
\mathbf{P} = \mathbf{K}\begin{bmatrix} \mathbf{R} & \mathbf{t} \end{bmatrix}
$$
where $\mathbf{t}$ is a $3\times 1$ vector represents the camera location, $\mathbf{R}$ is a $3\times 3$ rotation matrix represents the orientation of the camera with respect to an absolute coordinate frame, and $\mathbf{K}$ is a $3\times 3$ upper triangular matrix called calibration matrix.

For consective video frame, the crosspondent pixel pairs $(\mathbf{u}, \mathbf{u'})$ have the following relationship,
$$
\mathbf{u}' = \mathbf{K}'\mathbf{R}'\mathbf{R}^{-1}\mathbf{K}^{-1}\mathbf{u} \\
\mathbf{H} = \mathbf{K}'\mathbf{R}'\mathbf{R}^{-1}\mathbf{K}^{-1}
$$

If only pan or tilt rotation happend during this period, then the calibration matrix will be constant $\mathbf{K} = \mathbf{K}'$, thus,
$$
\mathbf{H}_{rot} = \mathbf{K}\mathbf{R}'\mathbf{R}^{-1}\mathbf{K}^{-1}
$$
Or if only zoom happend, then $\mathbf{R}' = \mathbf{R}$, thus,
$$
\mathbf{H}_{zoom} = \mathbf{K}'\mathbf{K}^{-1}
$$
