```python
tpl = cv2.imread("tpl.png", 0)
tplPts = cv2.findNonZero(tpl).reshape(-1, 2).astype(float)
initH = np.array([[  2.22268310e+00,  -3.21398359e-01,  -6.82516172e+02],
 [ -6.11967678e-03,   4.15528791e-01,   1.96014953e+02],
 [  5.86722019e-05,  -5.10694637e-04,   1.05304539e+00]])
```

```python
initP = cv2.perspectiveTransform(tplPts.reshape(-1,1,2), initH).reshape(-1,2)
cRot = cv2.warpPerspective(tpl, initH, (1280, 720))
fig, (ax1, ax2) = plt.subplots(1,2)
ax1.imshow(cRot, cmap='gray')
ax2.scatter(initP[:,0], initP[:,1], s=5, c='red', marker='.')
ax2.invert_yaxis(); ax2.set_aspect('equal')
#ax2.set_xlim([0,1280]); ax2.set_ylim([720, 0])
```
